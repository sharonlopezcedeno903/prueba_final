# Pregunta 9

# Pregunta 10

Para s√≥lo 'Costa Rica' y suponiendo que las emisiones se estancan a partir del √∫ltimo a√±o reportado (se mantiene igual). Cu√°ntas son las emisiones en 'Kg' s√≥lo para Costa Rica desde el ultimo a√±o registrado hasta el 2025 usando el dataframe 'df_emisions_co2'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
cr = df_emisions_co2[df_emisions_co2['Country'] == 'Costa Rica']

ultimo_periodo = cr['Year'].max()

emisiones_ultimo = cr[cr['Year'] == ultimo_periodo]['Value'].iloc[0]

periodos_faltantes = 2025 - ultimo_periodo

total_kilotoneladas = emisiones_ultimo * periodos_faltantes

total_kg = total_kilotoneladas * 1_000_000

print(f"Las emisiones totales en kg desde {ultimo_periodo} hasta 2025 para Costa Rica son: {total_kg}")


```

# Pregunta 12

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones totales en Centroam√©rica usando el dataframe 'df_emisions_co2'?
Se espera un dataframe de 2 columnas: 'Year' y 'total_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
paises_centroamerica = ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua']

filtro = df_emisions_co2[df_emisions_co2['Country'].isin(paises_centroamerica)]

reporte = filtro.groupby('Year')['Value'].sum().reset_index()

reporte.columns = ['Year', 'total_por_a√±o']

print(reporte)
```

# Pregunta 13

Para s√≥lo los pa√≠ses de Centroam√©rica ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua'] obtener un reporte por a√±o de las emisiones en promedio en Centroam√©rica usando el dataframe 'df_emisions_co2_per_capita'?
Se espera un dataframe de 2 columnas: 'Year' y 'promedio_por_a√±o'

```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
paises_centroamerica = ['El Salvador', 'Honduras', 'Guatemala', 'Costa Rica', 'Nicaragua']

filtro = df_emisions_co2_per_capita[df_emisions_co2_per_capita['Country'].isin(paises_centroamerica)]

reporte = filtro.groupby('Year')['Value'].mean().reset_index()

reporte.columns = ['Year', 'promedio_por_a√±o']

print(reporte)
```

# Pregunta 14

Cu√°l es el promedio de las emisiones per capita para el a√±o 2005 usando el dataframe 'df_emisions_co2_per_capita'?


```python
# Coloque el c√≥digo que responda a la pregunta aqui exactamente lo que se pregunta
promedio_2005 = df_emisions_co2_per_capita[df_emisions_co2_per_capita['Year'] == 2005]['value'].mean()

print(f"El promedio es de: {promedio_2005}")

```
